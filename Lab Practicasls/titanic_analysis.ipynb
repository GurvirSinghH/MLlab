# Titanic Dataset Analysis
This notebook performs exploratory data analysis (EDA) on the famous Titanic dataset.
We will load the dataset, perform data cleaning, and visualize some key insights.


```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for seaborn
sns.set_theme(style="whitegrid")
%matplotlib inline
```

```
# Load the Titanic dataset using seaborn
titanic = sns.load_dataset('titanic')

# Display the first few rows
titanic.head()
```

## Initial Exploration
Let's check the shape, information about columns, and basic statistics of the dataset.


```
print(f"Dataset Shape: {titanic.shape}\n")
print("Dataset Info:")
titanic.info()

print("\n\nBasic Statistics:")
display(titanic.describe(include='all'))
```

## Data Cleaning
We need to handle missing values before proceeding with the analysis. Let's see which columns have missing values.


```
# Check for missing values
titanic.isnull().sum()
```

```
# Handling missing values
# 'deck' has too many missing values, so we can drop it.
titanic = titanic.drop(columns=['deck'])

# 'age' can be imputed with the median age.
titanic['age'] = titanic['age'].fillna(titanic['age'].median())

# 'embarked' and 'embark_town' have very few missing values, we can drop those rows.
titanic = titanic.dropna(subset=['embarked', 'embark_town'])

# Verify that there are no more missing values
titanic.isnull().sum()
```

## Exploratory Data Analysis (EDA)
Let's explore the data to understand the factors that influenced survival.


```
# Plot the distribution of survival
plt.figure(figsize=(6, 4))
sns.countplot(data=titanic, x='survived', palette='Set1')
plt.title('Survival Distribution (0 = No, 1 = Yes)')
plt.show()
```

```
# Survival by Gender
plt.figure(figsize=(6, 4))
sns.countplot(data=titanic, x='survived', hue='sex', palette='Set2')
plt.title('Survival by Gender')
plt.show()
```

```
# Survival by Passenger Class
plt.figure(figsize=(6, 4))
sns.countplot(data=titanic, x='survived', hue='pclass', palette='Set3')
plt.title('Survival by Passenger Class')
plt.show()
```

```
# Age Distribution
plt.figure(figsize=(8, 5))
sns.histplot(data=titanic, x='age', kde=True, bins=30, color='purple')
plt.title('Age Distribution of Passengers')
plt.show()
```

```
# Survival by Age
plt.figure(figsize=(8, 5))
sns.histplot(data=titanic, x='age', hue='survived', kde=True, bins=30, palette='Set1', alpha=0.6)
plt.title('Age Distribution by Survival')
plt.show()
```

## Conclusion
- A significant portion of passengers did not survive.
- Females had a much higher survival rate compared to males.
- First-class passengers had a higher survival rate than those in second or third class.
- Children had a higher survival rate, while a large portion of young adults (around 20-30 years old) did not survive.
