## 1. Creating Dictionaries

```python
# Empty dictionary
empty = {}
print("Empty dict:", empty)

# Using curly braces
student = {"name": "Alice", "age": 21, "grade": "A"}
print("Student:", student)

# Using dict() constructor
car = dict(brand="Toyota", model="Corolla", year=2022)
print("Car:", car)

# From a list of tuples
pairs = [("x", 10), ("y", 20), ("z", 30)]
coords = dict(pairs)
print("Coords:", coords)

# Using dict.fromkeys()
keys   = ["a", "b", "c"]
default_dict = dict.fromkeys(keys, 0)
print("fromkeys():", default_dict)
```

## 2. Accessing Values

```python
person = {"name": "Bob", "age": 30, "city": "New York", "job": "Engineer"}
print("Dict:", person)

# Access by key
print("Name:", person["name"])

# get() — safe access, returns None (or default) if key missing
print("Age:", person.get("age"))
print("Country (missing):", person.get("country"))
print("Country (default):", person.get("country", "Unknown"))

# Access all keys, values, items
print("\nKeys:  ", list(person.keys()))
print("Values:", list(person.values()))
print("Items: ", list(person.items()))
```

## 3. Adding & Updating Entries

```python
inventory = {"apples": 10, "bananas": 5}
print("Original:", inventory)

# Add a new key-value pair
inventory["oranges"] = 8
print("After adding oranges:", inventory)

# Update existing value
inventory["apples"] = 20
print("After updating apples:", inventory)

# update() — merge another dict or keyword args
inventory.update({"grapes": 15, "bananas": 12})
print("After update():", inventory)

# setdefault() — set only if key doesn't exist
inventory.setdefault("mangoes", 7)
inventory.setdefault("apples", 999)  # won't override existing
print("After setdefault():", inventory)
```

## 4. Removing Entries

```python
data = {"a": 1, "b": 2, "c": 3, "d": 4, "e": 5}
print("Original:", data)

# del — remove by key
del data["a"]
print("After del 'a':", data)

# pop() — remove and return value
val = data.pop("b")
print(f"pop('b') returned {val}:", data)

# pop() with default (no KeyError if missing)
val2 = data.pop("z", "not found")
print(f"pop('z') returned: '{val2}'")

# popitem() — removes and returns last inserted key-value pair
last = data.popitem()
print(f"popitem() returned {last}:", data)

# clear() — empties the dictionary
temp = {"x": 1, "y": 2}
temp.clear()
print("After clear():", temp)
```

## 5. Iterating Over a Dictionary

```python
scores = {"Alice": 92, "Bob": 85, "Charlie": 78, "Diana": 95}

# Iterate over keys (default)
print("Keys:")
for key in scores:
    print(" ", key)

# Iterate over values
print("\nValues:")
for val in scores.values():
    print(" ", val)

# Iterate over key-value pairs
print("\nKey-Value pairs:")
for name, score in scores.items():
    print(f"  {name}: {score}")
```

## 7. Creating Dictionary From Lists

```python
#From 2 lists
keys = ['k1', 'k2', 'k3']
val = ['v1', 'v2', 'v3']
dict1 = dict(zip(keys, val))
print(dict1)

# from 4 lists
keys1 = ['a', 'b', 'c']
v1 = [1, 2, 3]
v2 = [4, 5, 6]
v3 = [7, 8, 9]

d = {k:(x, y, z) for k, x, y, z in zip(keys1, v1, v2, v3)}
print("Dictionary: ", d)
```
