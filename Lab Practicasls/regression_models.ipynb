# Linear and Logistic Regression Analysis
This notebook demonstrates how to build, evaluate, and visualize both Linear Regression and Logistic Regression models using `scikit-learn`.

We will use the **Diabetes dataset** for Linear Regression and the **Breast Cancer dataset** for Logistic Regression.


```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, confusion_matrix, classification_report
from sklearn.datasets import load_diabetes, load_breast_cancer

# Set style
sns.set_theme(style="whitegrid")
%matplotlib inline
```

## 1. Linear Regression
We will use the Diabetes dataset to predict disease progression based on several features. To visualize the linear regression, we will first train a model using a single feature, and then build a multiple linear regression model.


```
# Load Diabetes dataset
diabetes = load_diabetes()
X_lin = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
y_lin = diabetes.target

print(f"Features Shape: {X_lin.shape}")
X_lin.head()
```

```
from sklearn.linear_model import LinearRegression

# Let's use only one feature (BMI) for simple visualization
X_bmi = X_lin[['bmi']]

# Split the data
X_train_b, X_test_b, y_train_b, y_test_b = train_test_split(X_bmi, y_lin, test_size=0.2, random_state=42)

# Initialize and train the model
simple_lr = LinearRegression()
simple_lr.fit(X_train_b, y_train_b)

# Make predictions
y_pred_b = simple_lr.predict(X_test_b)

# Plotting the results
plt.figure(figsize=(8, 5))
plt.scatter(X_test_b, y_test_b, color='blue', label='Actual Data', alpha=0.6)
plt.plot(X_test_b, y_pred_b, color='red', linewidth=2, label='Regression Line')
plt.xlabel('BMI (Normalized)')
plt.ylabel('Disease Progression')
plt.title('Simple Linear Regression: BMI vs Disease Progression')
plt.legend()
plt.show()

print(f"Mean Squared Error: {mean_squared_error(y_test_b, y_pred_b):.2f}")
print(f"R2 Score: {r2_score(y_test_b, y_pred_b):.2f}")
```

```
# Now let's use all features
X_train_lin, X_test_lin, y_train_lin, y_test_lin = train_test_split(X_lin, y_lin, test_size=0.2, random_state=42)

multi_lr = LinearRegression()
multi_lr.fit(X_train_lin, y_train_lin)

y_pred_lin = multi_lr.predict(X_test_lin)

print("--- Multiple Linear Regression Results ---")
print(f"Mean Squared Error: {mean_squared_error(y_test_lin, y_pred_lin):.2f}")
print(f"R2 Score: {r2_score(y_test_lin, y_pred_lin):.2f}")

# Visualize Predicted vs Actual
plt.figure(figsize=(8, 5))
plt.scatter(y_test_lin, y_pred_lin, color='purple', alpha=0.6)
plt.plot([y_test_lin.min(), y_test_lin.max()], [y_test_lin.min(), y_test_lin.max()], 'r--', lw=2)
plt.xlabel('Actual Progression')
plt.ylabel('Predicted Progression')
plt.title('Actual vs Predicted Disease Progression (All Features)')
plt.show()
```

## 2. Logistic Regression
Logistic Regression is used for binary classification. We will use the Breast Cancer dataset to predict whether a tumor is malignant or benign.


```
# Load Breast Cancer dataset
cancer = load_breast_cancer()
X_log = pd.DataFrame(cancer.data, columns=cancer.feature_names)
y_log = cancer.target

# Target: 0 = Malignant, 1 = Benign
print(f"Features Shape: {X_log.shape}")
X_log.head()
```

```
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

# Split the data
X_train_log, X_test_log, y_train_log, y_test_log = train_test_split(X_log, y_log, test_size=0.2, random_state=42)

# It is good practice to scale features for Logistic Regression
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_log)
X_test_scaled = scaler.transform(X_test_log)

# Initialize and train the model
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train_scaled, y_train_log)

# Make predictions
y_pred_log = log_reg.predict(X_test_scaled)
```

```
# Evaluation
print(f"Accuracy Score: {accuracy_score(y_test_log, y_pred_log):.4f}\n")
print("Classification Report:")
print(classification_report(y_test_log, y_pred_log, target_names=cancer.target_names))

# Confusion Matrix
cm = confusion_matrix(y_test_log, y_pred_log)

plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=cancer.target_names, yticklabels=cancer.target_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix for Logistic Regression')
plt.show()
```
