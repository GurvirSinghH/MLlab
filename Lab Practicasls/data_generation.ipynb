## 1. Built-in `random` Module

```python
import random

print("--- Standard Python Random ---")
# Random float between 0.0 and 1.0
print("Random Float [0, 1):", random.random())

# Random integer within a range [a, b]
print("Random Integer [1, 10]:", random.randint(1, 10))

# Random float within a range
print("Random Uniform [5.5, 10.5]:", random.uniform(5.5, 10.5))

# Random choice from a sequence
colors = ["Red", "Green", "Blue", "Yellow"]
print("Random Choice:", random.choice(colors))

# Random sample (multiple unique items)
print("Random Sample (2 items):", random.sample(colors, 2))

# Shuffle a list in place
deck = [1, 2, 3, 4, 5]
random.shuffle(deck)
print("Shuffled List:", deck)
```

## 2. NumPy: Sequences & Distributions

```python
import numpy as np

# Set a seed for reproducibility
np.random.seed(42)

print("--- NumPy Random Arrays ---")
# Generate an array of random integers
ints = np.random.randint(low=10, high=50, size=5)
print("Array of Integers:", ints)

# Uniform distribution [0, 1) of specific shape
uniform_arr = np.random.rand(2, 3) # 2x3 matrix
print("\n2x3 Uniform Matrix:\n", uniform_arr.round(2))

# Normal (Gaussian) distribution: mean=0, std=1
normal_arr = np.random.randn(5)
print("\nStandard Normal Array:", normal_arr.round(2))

# Custom Normal distribution: mean=100, std=15
iq_scores = np.random.normal(loc=100, scale=15, size=5)
print("Custom Normal (IQ Scores):", iq_scores.round(1))

# Binomial distribution (e.g., 10 coin flips, p=0.5, repeated 5 times)
coin_flips = np.random.binomial(n=10, p=0.5, size=5)
print("Binomial (Heads out of 10 flips):", coin_flips)
```

## 3. Categorical & String Data Generation

```python
import string

n_samples = 10

# Generating categories with specific probabilities
categories = ["Electronics", "Clothing", "Home", "Toys"]
probs = [0.4, 0.3, 0.2, 0.1] # Must sum to 1.0
synth_cats = np.random.choice(categories, size=n_samples, p=probs)
print("Weighted Categories:", synth_cats)

# Generating random strings (e.g., Order IDs or Promo Codes)
def generate_codes(n, length=6):
    chars = string.ascii_uppercase + string.digits
    return [''.join(np.random.choice(list(chars), length)) for _ in range(n)]

promo_codes = generate_codes(5)
print("\nRandom Promo Codes:", promo_codes)
```

## 4. Time Series Data with Pandas

```python
import pandas as pd

print("--- Time Series Generation ---")
# Generate a sequence of dates (daily)
date_range = pd.date_range(start="2024-01-01", periods=7, freq='D')
print("Daily Date Range:\n", date_range)

# Generate random walks (useful for simulating stock prices)
steps = np.random.normal(loc=0, scale=1, size=7)
random_walk = np.cumsum(steps) # Cumulative sum of steps
start_price = 100
simulated_prices = start_price + random_walk

# Combine into a Series
ts_data = pd.Series(simulated_prices.round(2), index=date_range)
print("\nSimulated Stock Prices:\n", ts_data)
```

## 5. Building Synthetic DataFrames

```python
# Let's combine everything to create a realistic mock dataset
n_rows = 15

df_mock = pd.DataFrame({
    "user_id": range(1001, 1001 + n_rows),
    "join_date": pd.date_range(start="2023-01-01", periods=n_rows, freq='W'),
    "age": np.random.randint(18, 65, size=n_rows),
    "status": np.random.choice(["Active", "Inactive", "Pending"], size=n_rows, p=[0.7, 0.2, 0.1]),
    "monthly_spend": np.random.normal(loc=50, scale=15, size=n_rows).round(2),
    "satisfaction_score": np.random.randint(1, 6, size=n_rows)
})

# Introduce some missing values randomly for realism
missing_idx = np.random.choice(df_mock.index, size=3, replace=False)
df_mock.loc[missing_idx, "satisfaction_score"] = np.nan

print("Synthetic User Database:")
df_mock
```

## 6. Scikit-Learn: ML Dataset Generators

```python
from sklearn.datasets import make_classification, make_regression, make_blobs

print("--- Generating Data for Machine Learning ---")

# 1. Classification Data
# Generates features (X) and target labels (y)
X_clf, y_clf = make_classification(n_samples=5, n_features=4, n_classes=2, random_state=42)
print("\nClassification Features (X):\n", X_clf.round(2))
print("Classification Targets (y):", y_clf)

# 2. Regression Data
# Generates features and continuous target values
X_reg, y_reg = make_regression(n_samples=5, n_features=2, noise=0.1, random_state=42)
print("\nRegression Features (X):\n", X_reg.round(2))
print("Regression Targets (y):", y_reg.round(2))

# 3. Clustering Data (Blobs)
# Generates isotropic Gaussian blobs for clustering
X_blobs, y_blobs = make_blobs(n_samples=5, n_features=2, centers=3, random_state=42)
print("\nClustering Features (X):\n", X_blobs.round(2))
print("Cluster Assignments (y):", y_blobs)
```
